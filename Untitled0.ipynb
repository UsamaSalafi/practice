```
x = int(input("Input tables range: "))
if x >= 13:
  print("Note that table 15 is excluded")
  for t in range(13, x+1):
    if t==15:
      continue
    else:
      print("Starting The Table")
      for n in range(1,11):
        print(f"{t} * {n} = {t*n}")
else:
  print("Please input range value grater than 12")
```
